```python
import json
import numpy as np
import matplotlib.pyplot as plt

# Helper function to compute the duration of a video from the 'v' field
def compute_duration(video_uid):
    parts = video_uid.split('_')
    start_time = float(parts[1])
    end_time = float(parts[2])
    return end_time - start_time

# Helper function to process one JSON file and calculate performance for common videos
def process_json(file_path, common_video_ids, bins):
    with open(file_path) as f:
        data = json.load(f)

    # Initialize lists to store durations and whether predictions are correct
    durations = []
    correct_predictions = []

    # Loop through the data and extract relevant info
    for entry in data:
        video_uid = entry[1]['v']  # Ensure the correct index is used
        if video_uid in common_video_ids:  # Only process common videos
            duration = compute_duration(video_uid)
            durations.append(duration)

            # Store 1 for correct prediction, 0 otherwise
            correct_predictions.append(1 if entry[0]['pred'] == 'correct' else 0)

    # Initialize arrays to store correct/total counts for each bin
    binned_correct = np.zeros(len(bins)-1)
    binned_total = np.zeros(len(bins)-1)

    # Assign each duration to a bin and calculate correct/total predictions for each bin
    for i, duration in enumerate(durations):
        bin_idx = np.digitize(duration, bins) - 1  # Find the bin index
        if 0 <= bin_idx < len(binned_correct):
            binned_total[bin_idx] += 1
            binned_correct[bin_idx] += correct_predictions[i]

    # Calculate performance (correct/total) for each bin, avoid division by zero
    performance = np.divide(binned_correct, binned_total, out=np.zeros_like(binned_correct), where=binned_total != 0)
    return performance

# Function to get common video IDs from all three JSON files
def get_common_video_ids(file_paths):
    common_ids = None
    for file_path in file_paths:
        with open(file_path) as f:
            data = json.load(f)
            video_ids = {entry[1]['v'] for entry in data}  # Extract video IDs
            if common_ids is None:
                common_ids = video_ids
            else:
                common_ids.intersection_update(video_ids)  # Keep only common IDs
    return common_ids

# File paths to the three JSON files (replace these with your actual file paths)
file_paths = [
    '/content/drive/MyDrive/final_results_gemini_eval_new/Benchmark/qa_results_openQA_uniform_sampling_8_v4.json/eval_results.json',
    '/content/drive/MyDrive/final_results_gemini_eval_new/Benchmark/qa_results_openQA_uniform_sampling_32_v4.json/eval_results.json',
    '/content/drive/MyDrive/final_results_gemini_eval_new/Benchmark/qa_results_openQA_uniform_sampling_64_v4.json/eval_results.json'
]

# Define the bins for different duration ranges (e.g., 0-20, 20-40 seconds, etc.)
bins = np.arange(0, 100, 20)  # Adjusted to create 20-second ranges

# Get common video IDs present in all three files
common_video_ids = get_common_video_ids(file_paths)

# Process each file and calculate performance for common videos
performance1 = process_json(file_paths[0], common_video_ids, bins)
performance2 = process_json(file_paths[1], common_video_ids, bins)
performance3 = process_json(file_paths[2], common_video_ids, bins)

# Create a grouped bar plot
x = np.arange(len(bins) - 1)  # the label locations
width = 0.25  # the width of the bars

fig, ax = plt.subplots(figsize=(10, 6))

# Define a better color palette
colors = ['#1f77b4', '#ff7f0e', '#2ca02c']  # Blue, Orange, Green

# Plot the bars for each file with better colors
rects1 = ax.bar(x - width, performance1, width, label='8 frames', color=colors[0])
rects2 = ax.bar(x, performance2, width, label='32 frames', color=colors[1])
rects3 = ax.bar(x + width, performance3, width, label='64 frames', color=colors[2])

# Add labels and title
ax.set_xlabel('Video Duration Range (seconds)')
ax.set_ylabel('Performance (Correct/Total)')
ax.set_title('Prediction Performance by Video Duration Range for Common Videos')
ax.set_xticks(x)
ax.set_xticklabels([f'{int(bins[i])}-{int(bins[i+1])}' for i in range(len(bins)-1)])
ax.legend()

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import re
import json
import csv
from collections import defaultdict

def extract_prediction(prediction_str):
    # Use regex to find the letter inside brackets
    match = re.search(r'\[([A-Z])\]', prediction_str)
    return match.group(1) if match else None

def extract_prediction_answer(prediction_str):
    # Use regex to find the letter inside brackets
    match = re.search(r'\(([A-Z])\)', prediction_str)
    return match.group(1) if match else None

def compute_predictions(data):
    preds = []

    for item in data:
        # Extract the prediction
        prediction = extract_prediction(item['A'])

        # Extract the correct answer
        correct_answer = extract_prediction_answer(item['C'])

        preds.append({
            'video': item['V'],
            'category': item['M'],
            'question': item['QA'],  # Assuming 'Q' is the key for the question
            'prediction': prediction,
            'correct_answer': correct_answer,
            'is_correct': prediction == correct_answer
        })

    return preds

# Load the two JSON files
with open('/content/drive/MyDrive/final_results_gemini/qa_results_closeQA_uniform_sampling_flash_seed (1).json', 'rb') as f1:
    video_data1 = json.load(f1)

with open('/content/drive/MyDrive/final_results_gemini/qa_results_closeQA_uniform_sampling_in_context_learning_sorted_5ex.json', 'rb') as f2:
    video_data2 = json.load(f2)

# Compute predictions for both datasets
preds_1 = compute_predictions(video_data1)
preds_2 = compute_predictions(video_data2)

# Filter out entries where predictions are correct in one file and incorrect in the other
filtered_entries = []

for entry_1, entry_2 in zip(preds_1, preds_2):
    if entry_1['category'] == entry_2['category']:  # Ensure the categories are the same
        if entry_1['is_correct'] and not entry_2['is_correct']:
            filtered_entries.append({
                'video': entry_1['video'],
                'category': entry_1['category'],
                'question_1': entry_1['question'],
                'question_2': entry_2['question'],
                'prediction_1': entry_1['prediction'],
                'correct_answer_1': entry_1['correct_answer'],
                'correct_answer_2': entry_2['correct_answer'],
                'prediction_2': entry_2['prediction']
            })

# Save the filtered entries to a CSV file
csv_file_path = '/content/drive/MyDrive/final_results_gemini/filtered_results.csv'
with open(csv_file_path, 'w', newline='') as csvfile:
    fieldnames = ['video','category', 'question_1', 'question_2', 'correct_answer_1', 'correct_answer_2', 'prediction_1', 'prediction_2']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

    writer.writeheader()
    for entry in filtered_entries:
        writer.writerow({
            'video': entry['video'],
            'category': entry['category'],
            'question_1': entry['question_1'],
            'question_2': entry['question_2'],
            'correct_answer_1': entry['correct_answer_1'],
            'correct_answer_2': entry['correct_answer_2'],
            'prediction_1': entry['prediction_1'],
            'prediction_2': entry['prediction_2']
        })

# Print the results
print(f'Saved {len(filtered_entries)} filtered entries to CSV for comparison.')
```

```python
import os
import json
import argparse
import numpy as np
from PIL import Image, ImageDraw, ImageFont  # For combining text and images
import textwrap  # For text wrapping
import pandas as pd
from pytorchvideo.data.encoded_video import EncodedVideo
import matplotlib.pyplot as plt
import torch
import torchvision.io as io
from tqdm import tqdm
import cv2

def tensor_to_image(frame_tensor):
    """Convert a single frame tensor to a PIL image."""
    # Convert to byte range
    #frame_tensor = frame_tensor.clamp(0, 1).mul(255).byte()

    # Rearrange to HWC format
    frame_tensor = frame_tensor.permute(1, 2, 0)

    # Convert to numpy array
    frame_array = frame_tensor.cpu().numpy()

    # Return as a PIL image
    return Image.fromarray(np.uint8(frame_array))


def save_tensor_to_video(tensor, frame_rate=30):
    """
    Convert a tensor into an MP4 video and save it to a temporary file.

    Args:
    tensor (torch.Tensor): The video tensor in shape (frames, height, width, channels).
    frame_rate (int): The frame rate for the output video.

    Returns:
    str: The path to the saved video file.
    """
    # Convert tensor to numpy array
    tensor_np = tensor.permute(1, 2, 3, 0).cpu().numpy()  # Reorder to (frames, height, width, channels)
    if tensor_np.dtype != np.uint8:
      tensor_np = tensor_np.astype(np.uint8)
    # Get video dimensions
    num_frames, height, width, channels = tensor_np.shape

    # Create a temporary file
    temp_file = tempfile.NamedTemporaryFile(suffix='.mp4', delete=False)
    video_path = temp_file.name

    # VideoWriter setup
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # For .mp4 format
    out = cv2.VideoWriter(video_path, fourcc, frame_rate, (width, height))

    # Write frames to video
    for frame in tensor_np:
        # Convert frame to BGR for OpenCV
        frame_bgr = cv2.cvtColor(frame, cv2.COLOR_RGB2BGR)
        out.write(frame_bgr)

    out.release()  # Release the video writer

    return video_path

def save_frames_and_text(question, tensor1, output_text1, result_text1, category, response, name, dir_name='outputs_videollava_open/'):
    """Combines 2 rows of frames and texts into a single image."""
    def create_text_image(text, wrap_width=40):
      """Creates an image with the given text, wrapped to avoid overflow."""
      lines = text.split("\n")  # Split text by newline characters
      wrapped_lines = [textwrap.fill(line, wrap_width) for line in lines]  # Wrap each line individually
      wrapped_text = "\n".join(wrapped_lines)  # Join wrapped lines back into a single string
      text_image = Image.new("RGB", (400, 200), "white")  # Create image
      draw = ImageDraw.Draw(text_image)
      try:
        font = ImageFont.truetype("arial.ttf", 13)
      except IOError:
          font = ImageFont.load_default()
      draw.text((10, 10), wrapped_text, font=font, fill="black")  # Add wrapped text

      return text_image

    def prepare_frames_and_text(tensor, question, output_text, result_text, category, response):
        """Prepares frames and text images for a tensor."""
        tensor = tensor.permute(3, 0, 1, 2)
        num_channels, num_frames, height, width = tensor.shape
        # Choose three representative frames
        #torch.Size([850, 1080, 1440, 3])

        frame_indices = [0, num_frames // 2, num_frames - 1] if num_frames > 1 else [0] * 3
        frames = [tensor_to_image(tensor[:, idx, :, :]) for idx in frame_indices]
        position = question.lower().find("question")
        if position != -1:
            question= question[position:]

        # Combine text into one string
        final_text = f"{question}\nGT: {output_text}\nResult: {result_text}\nCat: {category}\nAcc: {response}"

        return frames, create_text_image(final_text)

    # Prepare frames and text for each tensor
    frames1, text_image1 = prepare_frames_and_text(tensor1, question, output_text1, result_text1, category, response)

    # Create a two-row image composition
    fig, axs = plt.subplots(1, 4, figsize=(20, 10))



    for i, frame1 in enumerate(frames1):
        axs[i].imshow(frame1)
        axs[i].axis('off')

    axs[3].imshow(text_image1)

    axs[3].axis('off')

    plt.tight_layout()
    directory_path = "/content/drive/MyDrive/"+dir_name
    if not os.path.exists(directory_path):
      # Create the directory
      os.makedirs(directory_path)

    plt.savefig(directory_path+name+".png")


def evaluate_answer(correct_answer, given_answer):
    # Extract the letter from the correct answer, assuming the format "(Letter) description"
    correct_letter = correct_answer[1]  # Extracts the letter assuming it's the second character in the string like '(C)'
    given_answer = given_answer[0]
    # Compare the extracted letter to the given answer
    return given_answer == correct_letter

def load_video_to_tensor(video_path):
    # Load video using torchvision.io
    video, audio, info = io.read_video(video_path, pts_unit='sec')
    return video, info



proper_names = {
    "LLM_Judge_evaluation": 'LLM_Judge_evaluation',
    "interpretation_of_social_context": 'Interpretation of social context',
    "non_existent_actions_with_existent_scene_depictions": 'Non-existent actions with existent scene depictions',
    'multiple_actions_in_a_single_video': 'Multiple actions in a single video',
    'unusual_and_physically_anomalous_activities': 'Unusual and Physically Anomalous activities',
    'partial_actions': 'Partial actions',
    'understanding_emotional_context': 'Understanding of emotional context',
    'fine_grained_action_understanding': 'Fine-grained action understanding',
    "interpretation_of_visual_context": 'Interpretation of visual context',
    'continuity_and_object_instance_count': "Continuity and Object Instance Count",
    "time_order_understanding": 'Time order understanding'
}


def parse_args():
    parser = argparse.ArgumentParser(description="question-answer-generation-using-gpt-3")
    parser.add_argument("--judge_results_path", required=True,
                        help="The path to the folder containing evaluation results from Judge LLM.")
    args = parser.parse_args()
    return args


def parse_results(judge_results_path, model):
    all_json_files = os.listdir(judge_results_path)
    category_results = {}
    import random
    random.shuffle(all_json_files)
    # Iterate through all JSON files
    for single_file_name in all_json_files:
        if 'json' not in single_file_name:
            continue
        with open(os.path.join(judge_results_path, single_file_name), "r") as json_file:
            contents = json.load(json_file)

        # Iterate through result_qa_pair to process each element
        for single_content, single_dict in contents:
            #print(single_content)
            #print(single_dict)
            cat = single_dict.get("cat")
            if type(cat)==list:
              cat = cat[0]
            cat = cat.lower()
            video_id = single_dict.get("v")
            answer = single_dict.get("a")
            outputs = single_dict.get("pred")
            question = single_dict.get("q")

            if cat is None:
                continue

            if cat not in category_results.keys():
                category_results[cat] = {"score_sum": 0, "count": 0, "yes_count": 0, "no_count": 0}

            # Computing score
            category_results[cat]["count"] += 1
            print(single_content)
            try:
                score_match = single_content['score']
            except:
                continue
                score_match = single_content[0]['score']

            score = int(score_match)
            category_results[cat]["score_sum"] += score

            # Computing accuracy
            try:
                pred = single_content['pred']
            except:
                try:
                    pred = single_content[0][0]['pred']
                except:
                    #continue
                    pred = single_content[0]['type']

            if pred.lower() == "correct":
                category_results[cat]["yes_count"] += 1
            elif pred.lower() == "incorrect":
                category_results[cat]["no_count"] += 1



            else:
                print(f"Error occurred during counting predictions for category {cat}")
            '''
            # Paths
            video_path = f'/content/drive/MyDrive/clips_ego4d_gemini/{video_id}.mp4'

            # Load the video into a tensor
            video_tensor, video_info = load_video_to_tensor(video_path)

            result_dict=(video_tensor, (answer).lower(), outputs)
            save_frames_and_text(question, result_dict[0], result_dict[1], result_dict[2], cat, pred.lower(), video_id, 'output_'+model+'_gemini_corrected/')
            '''
    categories = []
    accuracies = []
    counts = []

    import matplotlib.pyplot as plt

    # Compute and print the results for each category
    print(category_results)
    for category, results in category_results.items():
        if results["count"] > 0:
            average_score = results["score_sum"] / results["count"]
            accuracy = results["yes_count"] / (results["yes_count"] + results["no_count"])
            categories.append(category)
            accuracies.append(accuracy * 100)
            counts.append(results["count"])
            print(f"Accuracy for category [{category}]: {accuracy * 100:.2f}")
            print(f"Average score for category [{category}]: {average_score:.2f}")
        else:
            print(f"No data available for category [{category}]")
    # Plotting the counts
    plt.figure(figsize=(10, 6))
    plt.bar(categories, counts, color='tab:blue', alpha=0.6)
    plt.xlabel('Categories')
    plt.ylabel('Total Count')
    plt.title('Total Count for Each Category')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

    return categories, accuracies



# Parse the results from the five input files
categories_gemini, accuracies_gemini = parse_results('/content/drive/MyDrive/final_results_gemini_eval_new/EgoSchema/EgoSchema_results_closeQA_uniform_sampling_flash_seed.json', 'gemini')
categories_new, accuracies_new = parse_results('/content/drive/MyDrive/final_results_gemini_eval_new/EgoSchema/EgoSchema_results_closeQA_uniform_sampling_in_context_learning_sorted.json', 'gemini')
categories_new_1, accuracies_new_1 = parse_results('/content/drive/MyDrive/final_results_gemini_eval_new/GroundingDINO_videoSAM_text_objects_0.1.json', 'gemini')
categories_new_2, accuracies_new_2 = parse_results('/content/drive/MyDrive/final_results_gemini_eval_new/GroundingDINO_videoSAM_text_objects_hands.json', 'gemini')

# Create a set of all unique categories including the new datasets
all_categories = sorted( set(categories_gemini) | set(categories_new) | set(categories_new_1) | set(categories_new_2))

# Create dictionaries for fast lookup
gemini_dict = dict(zip(categories_gemini, accuracies_gemini))
new_dict = dict(zip(categories_new, accuracies_new))
new_1_dict = dict(zip(categories_new_1, accuracies_new_1))
new_2_dict = dict(zip(categories_new_2, accuracies_new_2))

# Align the accuracies based on the combined set of categories
aligned_accuracies_gemini = [gemini_dict.get(category, 0) for category in all_categories]
aligned_accuracies_new = [new_dict.get(category, 0) for category in all_categories]
aligned_accuracies_new_1 = [new_1_dict.get(category, 0) for category in all_categories]
aligned_accuracies_new_2 = [new_2_dict.get(category, 0) for category in all_categories]

# Plotting the grouped bar plot
x = np.arange(len(all_categories))  # the label locations
width = 0.15  # the width of the bars

fig, ax = plt.subplots(figsize=(14, 8))
# Plot the bars
rects2 = ax.bar(x - width, aligned_accuracies_gemini, width, label='CloseQA', color='orange')
rects3 = ax.bar(x, aligned_accuracies_new, width, label='VideoSAM - prompt 3', color='green')
rects4 = ax.bar(x + width, aligned_accuracies_new_1, width, label='VideoSAM - prompt', color='purple')
rects4 = ax.bar(x + 2*width, aligned_accuracies_new_2, width, label='VideoSAM - prompt 2', color='red')


# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_xlabel('Categories', fontsize=14)
ax.set_ylabel('Accuracy (%)', fontsize=14)
ax.set_title('Accuracy for Each Category', fontsize=16)
ax.set_xticks(x)
ax.set_xticklabels(all_categories, rotation=45, ha='right', fontsize=12)
ax.legend(fontsize=12)
ax.set_ylim([0, 100])

fig.tight_layout()

output_plot_path = '/content/drive/MyDrive/accuracy_comparison_plot_high_res.png'
plt.savefig(output_plot_path, dpi=300, bbox_inches='tight')
plt.show()
```
